= : Assign value
+= : Add and assign
-= : Subtract and assign

# Identity operators
is : Returns True if both variables are the same object

# Membership operators
in : Returns True if a sequence with the specified value is present in the object
not in : Returns True if a sequence with the specified value is not present in the object

# Precedence of operators
1. Parentheses
2. Exponentiation
3. Multiplication, Division, Floor division, Modulus
4. Addition, Subtraction

Pneumonic: PEMDAS (Parentheses, Exponents, Multiplication and Division, Addition and Subtraction)

## Example 1:

The example below shows how to use the above operators to perform arithmetic operations
on our calculator.

````python
expression = "hdhdhdh"

# This can be evaluated using the eval() function
# Same as: (2 + 6) - ((8 * 2) / 4) which is 4.0

try:
    result = eval(expression)
    print(f"The result of {expression} is {result}")
except ZeroDivisionError:
    print("You can't divide by zero!")
except NameError:
    print("You entered an invalid expression!")
except:
    print("An error occurred!")
````

# Simple math operations

```python
# Addition
print(2 + 3) # Output: 5
print(3 + 3.5) # Output: 6.5

# Subtraction
print(3 - 2) # Output: 1
print(3 - 2.5) # Output: 0.5
printf(2 - 3) # Output: -1

# Multiplication
print(2 * 3) # Output: 6
print(3 * 3.5) # Output: 10.5

# Division
print(3 / 2) # Output: 1.5
print(3 / 2.0) # Output: 1.5

# Floor division means rounding down and discarding the remainder
print(9 // 2) # Output: 4
print(7 // 2) # Output: 3

# Modulus
# Modulus operator returns the remainder of the division
print(9 % 2) # Output: 1
print(7 % 2) # Output: 1
print(6 % 2) # Output: 0

# Any value n % 2 will return 0 if n is even and 1 if n is odd

# Exponentiation(raise to the power of)
print(2 ** 3) # Output: 8
print(3 ** 2) # Output: 9
print(5 ** 0) # Output: 1


# Negation
x=10
neg_x = -x
print(neg_x) # Output: -10


# Comparison operators
print(3 == 2) # Output: False

print(3 != 2) # Output: True

print(3 > 2) # Output: True

print(3 < 2) # Output: False

print(3 >= 2) # Output: True
